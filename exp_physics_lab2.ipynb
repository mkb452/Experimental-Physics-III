```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy import optimize
```

```python
#Getting beats frequencies
time1, run1= np.loadtxt ("beats1.csv", skiprows = 1, unpack = True, delimiter=",")
print (time1)
print (run1)
```

```python
plt.figure(figsize=(30, 6))
plt.scatter(time1, run1)
```

```python
def test_func(x, a, b):
    return a * np.sin(b * x)

params, params_covariance = optimize.curve_fit(test_func, time1, run1,p0=[2, 2])
print(params)
print(params_covariance)
```

```python
plt.figure(figsize=(30, 6))
plt.scatter(time1, run1, label='Data')
plt.plot(time1, test_func(time1, params[0], params[1]),label='Fitted function')
plt.legend(loc='best')

plt.show()
```
