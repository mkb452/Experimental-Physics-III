```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#Simple Pendulum Content
run1_200, run2_200, run3_200= np.loadtxt ("simple_pendulum.csv", skiprows = 1, unpack = True, delimiter=",")
#Print arrays
print ("Run 1 =","\n", run1_200, "\n")
print ("Run 2 =", "\n",run2_200, "\n")
print ("Run 3 =", "\n",run3_200, "\n")
```

```python
run1avg_200 = np.average (run1_200,axis=None)
run2avg_200 = np.average (run2_200,axis=None)
run3avg_200 = np.average (run3_200,axis=None)
```

```python
print ("Run 1 average =",format(run1avg_200, ".2f"))
print ("Run 2 average =",format(run2avg_200, ".2f"))
print ("Run 3 average =",format(run3avg_200, ".2f"))
```
