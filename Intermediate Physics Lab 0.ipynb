```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#Print title
print ("Exercise One", "\n")
```

```python
#Create four arrays from text file given
tempData, tempError, pressData, pressError = np.loadtxt ("lab0_data.csv", skiprows = 1, unpack = True, delimiter=" ,")
print ("Below are the four different arrays:", "\n")
```

```python
#Print arrays
print ("Temperature =","\n", tempData, "\n")
print ("Temperature Error =", "\n",tempError, "\n")
print ("Pressure =", "\n",pressData, "\n")
print ("Pressure Error =", "\n",pressError, "\n")
```

```python
#Create Plot
plt.figure(1, figsize = (6,4) )
plt.plot(tempData,pressData,'b-',label="data")
plt.errorbar(tempData, pressData, fmt='ro', label="error",
xerr=tempError, yerr=pressError, ecolor='black')
plt.xlabel("T (C)")
plt.ylabel("P (lb/$in^2$)")
plt.legend(loc='upper right')
plt.show()
#Couldn't figure out why there were so many lines for this plot?
```

```python
#Exercise 2
#Print title
print ("Exercise Two", "\n")
```

```python
#Least squares fit to plot
slope,intercept,r_value,p_value,err = linregress(tempData, pressData)
plt.xlabel("T (C)")
plt.ylabel("P (lb/$in^2$)")
xfit = np.linspace(0, 300, 100) 
yfit = slope*xfit + intercept
plt.plot(xfit,yfit, 'r-', label='Linear Fit')
plt.legend(loc='best')
```

```python
#Linear chi-squared fit
slope,intercept,r_value,p_value,err = linregress(tempData, pressData)
plt.errorbar(tempData, pressData, fmt='o',label='Chi-squared weighting',yerr=pressError)
plt.xlabel("T (C)")
plt.ylabel("P (lb/$in^2$)")
xfit = np.linspace(0, 300, 100) 
yfit = slope*xfit + intercept
plt.plot(xfit,yfit, 'r-', label='Linear Fit')
plt.plot(tempData,pressData,'b-',label="data")
plt.legend(loc='best')
#Not sure why line stops in the middle
```
